# 5.3.4*: Currents in a Plane Wave

The problem asks to demonstrate that a plane wave with constant energy has independent left- and right-flowing probability current.  Specifically:

$$
  \psi(x) = Ae^{\I p x/\hbar} + B e^{-\I p x/\hbar}.
$$

From Eq. (5.3.8) we have the probability density current:

\begin{equation}
  \vect{j} = \frac{\hbar}{2m\I}(
    \psi^*\vect{\nabla}\psi 
    - \psi\vect{\nabla}\psi^* )
    =
    \frac{
      \braket{\psi|x}\braket{x|\op{P}|\psi}
      +
      \text{h.c}}{2m}.
    \tag{5.3.8}
\end{equation}

*(Note: in the alternative formulation, there is no integration over $x$: we have not inserted $\mat{1}$ here: this is the definition of the local current which depends on the position $x$.)*

First we simplify the expression a little, noting that the second term is just the hermitian conjugate of the first, and then noting that this will simply extract the real portion of the result with a factor of $1/2$:

$$
  \vect{j} = \frac{\hbar}{2m\I}\psi^*\vect{\nabla}\psi + \text{h.c.}
  = \Re\left(\frac{\hbar}{m\I}\psi^*\vect{\nabla}\psi\right)
  = \Im\left(\frac{\hbar}{m}\psi^*\vect{\nabla}\psi\right).
$$

I.e. let $z=a+b\I$ where $a$ and $b$ are real:

$$
  \frac{z - z^*}{2\I} = \frac{a + b\I - (a - b\I)}{2\I} 
  = \frac{2b\I}{2\I} = b = \Re\frac{z}{\I} = \Im z.
$$

Inserting the state into (5.3.8) we have

$$
  \vect{j} = \frac{\hbar}{2m\I}
    \left(A^*e^{-\I p x/\hbar} + B^*e^{\I p x/\hbar}\right)
    \frac{\I p}{\hbar}
    \left(Ae^{\I p x/\hbar} - Be^{-\I p x/\hbar}\right)
    +
    \text{h.c}
  = \frac{p}{2m}
    \left(A^*e^{-\I p x/\hbar} + B^*e^{\I p x/\hbar}\right)
    \left(Ae^{\I p x/\hbar} - Be^{-\I p x/\hbar}\right)
    +
    \text{h.c}
  = \frac{p}{2m}
    \left(\abs{A}^2 + AB^*e^{\I p x/\hbar} - A^*Be^{-\I p x/\hbar} - \abs{B}^2\right)
    +
    \text{h.c}
 = \frac{\abs{A}^2 - \abs{B}^2}{m}p
   +\frac{p}{2m}
    \left(
      AB^*e^{\I p x/\hbar}
      - A^*Be^{-\I p x/\hbar}
      +
      A^*Be^{-\I p x/\hbar}
      - AB^*e^{\I p x/\hbar}
    \right)
   = \left(\abs{A}^2 - \abs{B}^2\right)\frac{p}{m}.
$$

Thus, we may simply consider the $Ae^{\I x p/\hbar}$ piece as containing the right-flowing current and $Be^{-\I x p/\hbar}$ as the left-flowing current.  The cross-terms cancel, so we can form linear superpositions of right- and left-flowing currents in our scattering problems and they will behave independently. 

# 5.4.2*: Scattering


## (a) Delta Function

$$
  V(x) = V_0 a \delta(x).
$$

Everywhere except at $x=0$ the wavefunction must satisfy the free-particle Schrödinger equation, and so must be a sum of plane waves with energy $E = \hbar^2k^2/2m$.  Thus, we define the scattering wavefunctions as

$$
  \psi(x) = \begin{cases}
    \psi_L(x) = A e^{\I k x} + B e^{-\I k x} & x < 0\\
    \psi_R(x) = C e^{\I k x} + D e^{-\I k x} & 0 < x.
  \end{cases}
$$

The problem is solved by matching boundary conditions.  The wavefunction must be continuous, but the derivative can have a kink that is found by integrating the Schrödinger equation across the potential: 

$$
  \psi_L(0) = \psi_R(0), \qquad
  \psi_R'(0) - \psi_L'(0) = \int_{-\epsilon}^{\epsilon}\frac{2mV_0 a \delta(x)}{-\hbar^2}\psi(x)\d{x}
  = \frac{-2mV_0 a }{\hbar^2}\\
  A+B = C+D, \qquad
  \I k(A-B -C+D) = \frac{-2mV_0 a }{\hbar^2}(A+B).
$$

The transmission and reflection coefficients can be obtained by setting $A=1$, $B=r$, $C=t$, and $D=0$.  This simple substitution works because $k$ is the same on both left and right sides (otherwise one would need factors of $k_1/k_2$ to match the currents) and the last condition says there is no current flowing in from the right.  Solving, we find:

$$
  1+r = t, \qquad
  \I k(1-r - t) = -2r\I k = \frac{-2mV_0 a }{\hbar^2}(1+r),\\
  \frac{r}{1+r} = \frac{1}{r^{-1}+1} =\frac{mV_0 a }{\I\hbar^2 k} = \frac{1}{\I\beta},\\ 
  r = \frac{1}{\I\beta - 1}, \qquad
  \beta = \frac{\hbar^2 k^2}{m V_0}\frac{1}{ka}
$$

Hence:

$$
  R = \abs{r}^2 = \frac{1}{1+\beta^2}, \qquad
  T = 1 - R = \frac{1}{1+\beta^2}.
$$

### Comments

Note that the reflection coefficient $R \rightarrow 1$ goes to one when $\beta^2$ becomes large – i.e. when $k^2 \propto E$ is small.  Thus, low energy particles are always reflected from sharp potentials.  Second, notice that the reflection coefficient depends on $(aV_0)^2$, so reflection occurs *even if the potential is attractive*!  This is a purely quantum effect – classical particles will pass through an attractive barrier.

## (b) Square Well


We now consider a finite square-well potential with width $2a$:

$$
  V(x) = \begin{cases}
    V_0 & \abs{x}\leq a\\
    0 & \abs{x} \geq a.
  \end{cases}.
$$

### Think

In the limit $a\rightarrow 0$ holding $2aV_0$ constant, we should reroduce the results of part (a) (with $a->2a$).

### Strategy

In each of the three regions of space, the potential is flat, so we can know the form of the wavefunction since it must satisfy the Schrödinger equation (see Exercise (5.1.2) from the last assignment).  A brute force solution would be to set:

\begin{align}
  x &< -a, &
  \psi_{I}(x) &= A e^{\I k x} + B e^{-\I k x}, &
  \hbar k &= \sqrt{2mE}, \tag{I}\\
  -a \leq &x < a, &
  \psi_{II}(x) &= \alpha e^{-\kappa x} + \beta e^{\kappa x}, &
  \hbar \kappa &= \sqrt{-2m(E-V_0)}, \tag{II}\\
  a &< x, &
  \psi_{III}(x) &= C e^{\I k x} + D e^{-\I k x}, &
  \hbar k &= \sqrt{2mE}. \tag{III}
\end{align}

The strategy will now be to make sure that the wavefunction and its derivative are continuous everywhere - meaning at the boundaries $x = \pm a$.  This should give us some equations we can use to solve for the energy eigenstates.

\begin{align}
  \psi_{I}(-a) &= \psi_{II}(-a), & \psi_{II}(a) &= \psi_{III}(a), \\
  \psi_{I}'(-a) &= \psi_{II}'(-a), & \psi_{II}'(a) &= \psi_{III}'(a).
\end{align}

As before, since $V(x>\abs{a})=0$, we can solve for the reflection and transmission coefficients by setting $A=1$, $B=r$, $C=t$, and $D=0$. 
We have 4 unknowns $r$, $t$, $\alpha$,  and $\beta$, and four equations. Thus, we are ready to proceed.

### Details

Plugging in our ansatz and being careful with the signs, we obtain the following equations:

\begin{align}
  Ae^{-\I ka} + Be^{\I ka} 
  &= \alpha e^{\kappa a} + \beta e^{-\kappa a},\\
  \frac{\I k}{\kappa}(Ae^{-\I ka} - Be^{\I ka})
  &= -\alpha e^{\kappa a} + \beta e^{-\kappa a},\\
  Ce^{\I ka} + De^{-\I ka} 
  &= \alpha e^{-\kappa a} + \beta e^{\kappa a},\\
  \frac{\I k}{\kappa}(Ce^{\I ka} - De^{-\I ka})
  &= -\alpha e^{-\kappa a} + \beta e^{\kappa a}.
\end{align}

Solving for $\alpha$ and $\beta$ in each set of equations, then eliminating gives:

\begin{align}
  \alpha &= \frac{Ae^{-\I ka} + Be^{\I ka} - \frac{\I k}{\kappa}(Ae^{-\I ka} - Be^{\I ka})}{2e^{\kappa a}}
  = \frac{Ce^{\I ka} + De^{-\I ka} - \frac{\I k}{\kappa}(Ce^{\I ka} - De^{-\I ka})}{2e^{-\kappa a}},\\ 
  \beta &= \frac{Ae^{-\I ka} + Be^{\I ka} + \frac{\I k}{\kappa}(Ae^{-\I ka} - Be^{\I ka})}{2e^{-\kappa a}}
  = \frac{Ce^{\I ka} + De^{-\I ka} + \frac{\I k}{\kappa}(Ce^{\I ka} - De^{-\I ka})}{2e^{\kappa a}}.
\end{align}

Setting $A=1$, $B=r$, $C=t$, and $D=0$, and solving, we have:

\begin{align}
  \frac{\left(1 - \frac{\I k}{\kappa}\right)e^{-\I ka} 
  + re^{\I ka}\left(1 + \frac{\I k}{\kappa}\right)}{e^{2\kappa a}}
  = te^{\I ka}\left(1 - \frac{\I k}{\kappa}\right),\\ 
  \frac{e^{-\I ka}\left(1 + \frac{\I k}{\kappa}\right)
  + re^{\I ka}\left(1 - \frac{\I k}{\kappa}\right)}{e^{-2\kappa a}}
  = te^{\I ka}\left(1 + \frac{\I k}{\kappa}\right).
\end{align}

To simplify, let $Q=1 + \frac{\I k}{\kappa}$ and $\tilde{Q}=1 - \frac{\I k}{\kappa}$.  *Note: Although $\tilde{Q}=Q^*$ for the problem as stated in the book, keeping these independent will allow us to extend our result to the case $E>V_0$ when this is not true.*:

$$
  te^{\I ka} 
  = \frac{\tilde{Q}e^{-\I ka} + re^{\I ka}Q}{\tilde{Q}e^{2\kappa a}}
  = \frac{e^{-\I ka}Q + re^{\I ka}\tilde{Q}}{Q e^{-2\kappa a}},\\
  r = \frac{e^{-2\I ka}\left(e^{2\kappa a}
     - e^{-2\kappa a}\right)}{\left(
    \frac{Q}{\tilde{Q}e^{2\kappa a}}
    -
    \frac{\tilde{Q}}{Q e^{-2\kappa a}}
  \right)}
  = \frac{2\sinh(2\kappa a)}{e^{2\I ka}\left(
    \frac{Q}{\tilde{Q}e^{2\kappa a}}
    -
    \frac{\tilde{Q}}{Q e^{-2\kappa a}}
  \right)},\\
  R = \abs{r}^2.
$$


## Comments

A key to working though this type of algebra is to keep everything as symmetric as possible, and then to look for common sub-expressions like $Q$ which you can use to express your answer in a simple form.

As we mentioned before, we expect that in the limit of small $a$, the result should approach that of a $V(x)= 2aV_0\delta(x)$ potential:

$$
  R_{\delta} = \frac{1}{1+\beta^2}, \qquad
  \beta = \frac{\hbar^2 k^2}{2m V_0 ka} 
            = \frac{E}{V_0}\frac{1}{ka}.
$$

Our problem has as parameters, $E$, $a$, $V_0$, $\hbar$, and $m$. From these, we can form two dimensionless parammeters.  The combination $\beta$ above characterizes the wavelength $k$ of the incoming wave and is the only parameter governing the reflection from the delta-function potential. The other parameter is:

$$
  \nu = \frac{kaV_0}{\hbar^2/2ma^2} = \frac{2mka^3V_0}{\hbar^2}.
$$

This determines how small the width of the potential is.  For small $\nu$, we expect the potential to behave as a delta-function.

```python
%pylab inline --no-import-all
kaV0 = m = hbar = 1
nu = 1.0

def get_R(E, nu=nu):
  k = np.sqrt(2*m*E)/hbar
  a = np.sqrt(abs(nu)*hbar**2/kaV0/2/m)
  V_0 = np.sign(nu)*kaV0/k/a
  kappa = np.sqrt(-2*m*(E-V_0)+0j)/hbar
  Q = 1 + 1j*k/kappa
  Qt = 1 - 1j*k/kappa
  r = (2*np.sinh(2*kappa*a)
       / np.exp(2j*k*a)
       / (Q/Qt/np.exp(2*kappa*a) 
          - Qt/Q*np.exp(2*kappa*a)))
  return abs(r)**2

@np.vectorize
def get_R1(E, nu=nu):
  """This is a check from another technique..."""
  k = np.sqrt(2*m*E)/hbar
  a = np.sqrt(abs(nu)*hbar**2/kaV0/2/m)
  V_0 = np.sign(nu)*kaV0/k/a
  k1 = np.sqrt(2*m*(E-V_0)+0j)/hbar
  Q = 1. + k/k1
  Qt = 1. - k/k1
  T_step = np.array([[Q, Qt], [Qt, Q]])/2
  D = np.diag([np.exp(1j*2*a*k1), np.exp(-1j*2*a*k1)])
  
  T = np.linalg.solve(T_step, D.dot(T_step))
  S = np.array([[np.linalg.det(T), T[0,1]],
                [-T[1,0], 1]])/T[1, 1]
  return(abs(S[0,1])**2)

def get_R_delta(E):
  """Return the reflection coefficient of the
  corresponding delta-function potential."""
  k = np.sqrt(2*m*E)/hbar
  beta = E / kaV0
  return 1/(1+beta**2)


plt.figure(figsize=(10, 5))
E = np.linspace(1e-6, 3, 1000)
assert np.allclose(get_R(E), get_R1(E))
plt.plot(E, get_R_delta(E), ':', label=r'$\delta(x)$')
for nu in [0.01, 0.1, 0.5, 10, 100]:
  plt.plot(E, get_R(E, nu=nu), '-', 
           label=r"$\nu={}$".format(nu))
#for nu in [-0.1, -0.5, -1]:
#  plt.plot(E, get_R(E, nu=nu), '--', 
#           label=r"$\nu={}$".format(nu))
plt.legend()
plt.xlabel('$E/|V_0|ka$')
plt.ylabel('$R$')
```

We see that our analysis is correct: as $\nu$ gets small, our finite barrier scattering approaches the delta-function results.  Notice that there are some interesting features with the finite barrier where for special energies, the reflection coefficient almost vanishes.  What is special about these energies?  These occur where $E > V_0$ so that $\kappa = \I k_1$ and then the numerator of $R \propto \sinh(2\kappa a) \propto \sin(2ka)$, so these occur where $ka = \pi n$.

```python
%pylab inline --no-import-all
V0 = m = hbar = 1
nu = 1.0

def get_R(E, nu=nu):
  k = np.sqrt(2*m*E)/hbar
  a = (hbar**2*abs(nu)/2/m/k/V0)**(1./3.)
  kappa = np.sqrt(-2*m*(E-V0)+0j)/hbar
  Q = 1 + 1j*k/kappa
  Qt = 1 - 1j*k/kappa
  r = (2*np.sinh(2*kappa*a)
       / np.exp(2j*k*a)
       / (Q/Qt/np.exp(2*kappa*a) 
          - Qt/Q*np.exp(2*kappa*a)))
  return abs(r)**2

def get_R_delta(E, ka):
  """Return the reflection coefficient of the
  corresponding delta-function potential."""
  k = np.sqrt(2*m*E)/hbar
  beta = E / ka / V0
  return 1/(1+beta**2)

plt.figure(figsize=(10, 5))
E = np.linspace(1e-6, 10, 10000)
for nu in [0.01, 0.1, 0.5, 10, 100]:
  k = np.sqrt(2*m*E)/hbar
  a = (hbar**2*abs(nu)/2/m/k/V0)**(1./3.)
  x = E/V0
  l, = plt.plot(x, get_R(E, nu=nu), '-', 
                label=r"$\nu={}$".format(nu))
  plt.plot(x, get_R_delta(E, ka=k*a), ':', c=l.get_c())
#for nu in [-0.1, -0.5, -1]:
#  plt.plot(E, get_R(E, nu=nu), '--', 
#           label=r"$\nu={}$".format(nu))
plt.legend()
plt.xlabel('$E/|V_0|$')
plt.ylabel('$R$')
plt.xlim(0,3)
```

This is the same plot with a different scaling on the abscissa to show where $E=V_0$ (at 1).  On this plot, the delta-function is not constant.
